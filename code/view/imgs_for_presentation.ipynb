```python
image = sitk.ReadImage(r'Z:\grodriguez\CardiacOCT\data-original\extra-scans-DICOM\NLD-RADB-0007.dcm')
image_pixel_data = sitk.GetArrayFromImage(image)
```

```python
seg = sitk.ReadImage(r'Z:\grodriguez\CardiacOCT\data-original\extra-segmentations-ORIGINALS\NLD-RADB-0007.nii.gz')
seg_pixel_data = sitk.GetArrayFromImage(seg)
```

```python
fig, axes = plt.subplots(1,2,figsize=(15,15))

frame = 273

axes[0].imshow(image_pixel_data[frame,:,:,:])
axes[1].imshow(seg_pixel_data[frame,:,:])
```

```python
np.unique(seg_pixel_data[frame,:,:])
```

```python
_, rows, cols = seg_pixel_data.shape

only_label = np.zeros((rows, cols))

for i in range(rows):
    for j in range(cols):
        if seg_pixel_data[frame, i, j] == 10:
            only_label[i, j] = 1

        else:
            only_label[i, j] = 0
```
